```python
df = pd.read_csv('C:/Users/Ruben/Documents/GitHub/ReACT_GCV/code/work/scrape_environment/image_npg_1/txt/txt_csv_iter1.csv')
```

```python
sents = [ast.literal_eval(i) for i in df['txt']]
sents = [" ".join([" ".join(s) for s in par]) for par in sents]
sents = [s.split(' ') for s in sents]
ids = [i.split('_')[1] for i in list(df['id'])]
sents = zip(ids,sents)
```

```python
documents = [TaggedDocument(doc, [ids]) for ids, doc in sents]
```

```python
d2v_model = Doc2Vec(documents, vector_size=150, window=12, min_count=1, workers=4)
```

```python
vec1 = d2v_model.infer_vector(ids[0])
vec2 = d2v_model.infer_vector(ids[1])

print(spatial.distance.cosine(vec1, vec2))
```

```python
for i in ids:
    
    for i2 in ids:
        
        if i != i2:
            vec1 = d2v_model.infer_vector(i)
            vec2 = d2v_model.infer_vector(i2)
            sim = 1 - spatial.distance.cosine(vec1, vec2)
            
            if sim > 0.8:
                print(i,i2)
```
