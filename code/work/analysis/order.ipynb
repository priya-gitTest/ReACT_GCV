```python
from tqdm import tqdm
import os
import editdistance
base_path = "/media/ruben/Data Drive/react-data/protest/selection"
```

```python
c = []

for photo in [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]:
    photo_folder = os.path.join(base_path, photo, "context_images")
    ct = [os.path.join(photo_folder,f) for f in os.listdir(photo_folder) if ".txt" in f]
    c += ct
```

```python
d = dict()
    
for id_ in c:
    with open(id_,'r') as f:
        content = f.readlines()[0]
        url = content.split(': ')[1]
        
        if url not in d.keys():
            d.update({url:id_})
```

```python
selection = list(set(list(d.values())))
```
