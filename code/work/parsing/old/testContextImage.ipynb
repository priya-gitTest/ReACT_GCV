```python
from bs4 import BeautifulSoup
import requests
import image_scraper
from functions import *
from urllib.parse import urljoin
import json
import uuid
```

```python
# Configuration
extensions = ".jpg .JPG .JPEG .jpeg .Jpeg .png .bmp .ppm".split(' ')

def imgTag(url):
    try:
        content = requests.get(url,t).content
    except Exception as e:
        print(e)
        return
    soup = BeautifulSoup(content,'lxml')
    image_tags = []
    for tag in ['img','meta','a']:
        tt = soup.findAll(tag)
        image_tags = image_tags + tt
    
    list_url = []
    for c,tag in enumerate(image_tags):
        attributes = dict(tag.attrs)
        
        for k,v in attributes.items():
            
            # Extract Image
            if any(substring in v for substring in extensions):
                list_url.append(v)

    if len(list_url) > 0:
        return list_url
    else:
        return
```
