```python
def MoreCapitals(line):
    
    sum_strange = 0
    
    for word in line:
        sum_strange += sum(1 for c in word if c.isupper())
    return [sum_strange, len(line)]

def getLines(num):
    txt = df['text'][num]
    if str(txt) == "nan":
        return
    
    ## Split on Newlines
    lines = [i for i in txt.split('\r\n')]
    
    ## Tokenize Lines
    tokenized_lines = [[x for x in i.split(' ') if x] for i in lines]
    
    ## Select Lines with > 10 tokens:
    selection = [" ".join(l) for l in tokenized_lines if len(l) > 10]    
    
    ## Select Lines with 'Normal' Tokens (no words with > 1 capital letters)
    selection = [MoreCapitals(l) for l in selection]
    return selection
```
