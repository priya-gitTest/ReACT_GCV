```python
def getLines(num):
    txt = df['text'][num]
    if str(txt) == "nan":
        return
    lines = [i for i in txt.split('\r\n')]
    return [(l,c) for c,l in enumerate(lines)]
    
    #tokenized_lines = [[x for x in i.split(' ') if x] for i in lines]
    #tokenized_lines = [" ".join(x) for x in tokenized_lines]
    #selection = [" ".join(l) for l in tokenized_lines if len(l) > 10]    
    
```

```python
failed = 0
succeeded = 0
results = []
for i in range(len(df)-1):
    res = getLines(i)
    
    if res is None or len(res) == 0:
        failed += 1
    else:
        succeeded += 1
        results.append(res)
        
print("{} failed".format(failed))
print("{} success".format(succeeded))
```

```python
def TrafiLocal(file):
    try:
        with open(file + ".html", 'r', encoding="UTF-8") as infile:
            soup = BeautifulSoup(infile, "html.parser")
            c = str(soup.contents)
            r = trafilatura.extract(c)
            if r is not None:
                return r
            else:
                return
    except Exception as e:
        print(e)
        return

def TrafiWeb(url):
    res = trafilatura.fetch_url(url)
    res = trafilatura.extract(res)
    return res
```
